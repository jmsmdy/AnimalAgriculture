```python
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(figsize=(20,20))

US = gpd.read_file('US Counties/UScounties.shp')

minnesota = US[US['STATE_FIPS'] == '27']
minnesota['FIPS2'] = minnesota['FIPS'].apply(lambda x: int(x))
minnesota.plot('FIPS2',
               cmap='OrRd',
               ax=ax,
               legend=True,
               legend_kwds={'label': "Population by Country",
                            'orientation': "horizontal"})

ax.axis("off")
```
